```python
# import data
train=pd.read_csv('mnist_train.csv')
test=pd.read_csv('mnist_test.csv')
```

```python
# get important parameters
#k -calss
#n_obs of observations
#n_var
k=train.iloc[:,0].max()
n_obst = len(train)
n_var=train.shape[1]-1
# prepare for train
x_train=train.iloc[:,1:]
y_train=train.iloc[:,0]
x_test=test.iloc[:,1:]
y_test=test.iloc[:,0]
```

```python
# get probability matrix
def softmax(z):
# Normalize the scores beforehand with max as zero to avoid 
# numerical problems with the exponential
    exps=np.exp(z-np.max(z,axis=0,keepdims=True))
    s=exps/np.sum(exps,axis=0,keepdims=True)
    return s
```

```python
'''
get the derivative of softmax function
s is k*n matrix
x is n*d  matrix 
output k*d matrix 
k-class, d-# varaibles,n-# of cases
'''
def deriv_soft(s,x,y):
    n_obst=x.shape[0]
    # select the goal label and substrcat 1
    s[y-1,np.arange(n_obst)] -= 1
    return np.dot(s,x)
    
```

```python

   
def get_mini_batch(x_train,y_train,size=2**10):
    
    n_samples=x_train.shape[0]
    x_mini_batches=list()
    y_mini_batches=list()
    index=np.random.permutation(n_samples)
    x_shuffle=x_train.iloc[index,:]
    y_shuffle=y_train.iloc[index,]
    itr=round(n_samples/size)
    print(itr)

    for i in range(itr):
       
        x_mini=x_shuffle.iloc[i*size:(i+1)*size,:]
        y_mini=y_shuffle.iloc[i*size:(i+1)*size,]
        x_mini_batches.append(x_mini)
        y_mini_batches.append(y_mini)
        
    return x_mini_batches,y_mini_batches,itr
```

```python
def get_gd(x_train,y_train,w):
    b=x_train.shape[0]
    z=np.dot(x_train,w).T
    s=softmax(z)
    gd=deriv_soft(s,x_train,y_train)/b   
    return gd
```

```python
'''
itr=number of iteration for gradient dexcent
a is learning rate
size is size of mini bacth
itr_mini :number of mini batch to get average gd
'''
def multi_logistic_train(x_train, y_train, itre=100,a=0.05):
    x_minis,y_minis,itr=get_mini_batch(x_train,y_train)
    n_var=x_train.shape[1]
    # intintial w
    w=np.random.rand(k*n_var)
    w=w.reshape([n_var,k])
    for j in range(itre):
        for i in range(itr):
            x_mini=x_minis[i]
            y_mini=y_minis[i]
            # gradient descent
            gd=get_gd(x_mini,y_mini,w)
            w=w-a*gd.T     
    return(w)
```

```python
# train data
w=multi_logistic_train(x_train, y_train)
print(w.shape)
print(w)
```

```python
# predict
def multi_logistic_predict(x_test,w):
    z=np.dot(x_test,w)
    y_pred=np.argmax(z,axis=1)
    return y_pred
```

```python
def accuracy(y_test,y_pred):
    n_test=y_test.shape[0]
    correct=np.sum(y_test==y_pred)
    return correct/n_test

```

```python
#predict test dataset
y_pred=multi_logistic_predict(x_test,w)
print(y_pred)
```

```python
# report accuracy
accu=accuracy(y_test,y_pred)
print(accu)
```
